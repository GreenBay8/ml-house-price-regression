```python
import pandas as pd

data = {
    "LotArea": [8450, 9600, 11250, 9550, 14260],
    "OverallQual": [7, 6, 7, 7, 8],
    "YearBuilt": [2003, 1976, 2001, 1915, 2000],
    "GrLivArea": [1710, 1262, 1786, 1717, 2198],
    "BedroomAbvGr": [3, 3, 3, 3, 4],
    "SalePrice": [208500, 181500, 223500, 140000, 250000]
}

df = pd.DataFrame(data)
df.to_csv("data.csv", index=False)

print("data.csv created")

```

```python
df = pd.read_csv("data.csv")
print(df.shape)
df.head()
```

```python
# Quick summary of the dataset
print("Dataset info:")
df.info()

print("\nFirst 5 rows:")
df.head()

print("\nBasic statistics:")
df.describe()
```

```python
# Check for missing values
print(df.isnull().sum())

# Basic statistics
print(df.describe())
```

```python
import matplotlib.pyplot as plt

# Scatter plot: Living area vs Sale Price
plt.scatter(df['GrLivArea'], df['SalePrice'])
plt.xlabel('GrLivArea')
plt.ylabel('SalePrice')
plt.title('Living Area vs Sale Price')
plt.show()

```

```python
# Using GrLivArea as a single feature to demonstrate
# a simple end-to-end regression workflow

from sklearn.linear_model import LinearRegression

X = df[['GrLivArea']]  # feature
y = df['SalePrice']    # target

model = LinearRegression()
model.fit(X, y)


print(f"Sample prediction: {model.predict([[1500]])}")
```

```python
import pandas as pd

model.predict(pd.DataFrame([[1500]], columns=['GrLivArea']))
```
